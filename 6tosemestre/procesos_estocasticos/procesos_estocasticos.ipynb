La incorporoacion de $D_{ik}$ proporciona una motivacion para formular  un problema de programacion lineal .El inconveniente para utilizar este modelo  es que las $D_{ik}$ son valores enteros por lo que se sugiere la siguinte transformación

$$D_{ik}=P(decisión=k/estado=i)$$

Entonces ($D_{i1},D_{i2},...,D_{ik}$) es la distribución de la probabilidad para la decisión debra tomarse para el estado i. a este tipo de politicas se les conoce como aleatorizados 

La politica queda entonces determinada por porla matriz D , donde cadarenglon suma 1 y 

$$0\leq D_{ik}\leq1$$

un ejemplo para el caso de la maquina seria 

| |1|2|3|
|-|-|-|-|
|**0**|1|0|0|
|**1**|1/2|0|1/2|
|**2**|1/4|1/4|1/2|
|**3**|0|0|1|

Con este adecuación ya es posible platear el modelo de programación lineal.las variables de decisión $Y_{ik}$ se definen como:

$$Y_{ik}=P(estado=i y decisión=k)$$

$$i=0,1,..,M K=1,2,...,M$$

cada $Y_{ik}$ se relaciona con las $D_{ik}$ de la siguiente manera

$$Y_{ik}=\pi(i)D_{ik}$$

donde

$$\pi(i)=\sum\limits_{K=1}^{k}Y_{ik}$$

de manera que

$$D_{ik}=\frac{Y_{ik}}{\pi_i}=\frac{Y_{ik}}{\sum\limits_{K=1}^{k}Y_{ik}}$$

EXISTEN VARIAS RESTRICCIONES OBRE LAS $Y_{ik}$ 

1. $\sum\limits_{i=0}^{M}\pi(i)=1$ de manera que $\sum\limits_{i=0}^{M}\sum\limits_{k=1}^{k}Y_{ik}=1$

2. Sobre las probabilidades a laro plazo $\pi(i)=\sum\limits_{i=0}^{M}\pi(i)P_{ij}$

   de manera que $\sum\limits_{k=1}^{k}Y_{jk}=\sum\limits_{i=0}^{M}\sum\limits_{k=1}^{k}Y_{ik}P_{ij}(k)$ j=0,1,2..m
   
   
3. $Y_{ik}\geq0$ i=0,1,...M  K=1,2,...,M


El costo promedio esperado por unidad de tiempo  a larga es 

$$E(c)=\sum\limits_{i=0}^{M}\sum\limits_{k=1}^{k}\pi(i)C_{ik}D_{ik}$$

$$=\sum\limits_{i=0}^{M}\sum\limits_{k=1}^{k}C_{ik}Y_{ik}$$

entonces el ppl consiste en:

$$\sum\limits_{k=1}^{k}Y_{jk}-\sum\limits_{i=0}^{M}\sum\limits_{k=1}^{k}Y_{ik}P_{ij}(k)=0$$

$$Y_{jk}\geq0$$
