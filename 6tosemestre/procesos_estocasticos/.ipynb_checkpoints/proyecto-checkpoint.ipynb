```python
#rangos
min=20
max=40
```

```python
total=0
for x in range(0,len(data[0])):
    if data.iloc[x,0]>=min and data.iloc[x,0]<=max:
        total+=1
total
```

```python
notomados=0
for x in range(0,len(data[0])):
    if (data.iloc[x,5]==' Married-spouse-absent' or data.iloc[x,5]==' Married-AF-spouse') and (data.iloc[x,0]>=min and data.iloc[x,0]<=max):
        notomados+=1
notomados
```

```python
totalreal=total-notomados
totalreal
```
