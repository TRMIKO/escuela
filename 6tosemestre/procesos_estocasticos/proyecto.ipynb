```python
totales=np.sum(arr,axis=1)
trans=np.zeros(shape=(4,4))
for i in range(0,4):
    for j in range(0,4):
        if i == 0:
            trans[i][j]=arr[i][j]/totales[0]
        elif i == 1:
            trans[i][j]=arr[i][j]/totales[1]
        elif i == 2:
            trans[i][j]=arr[i][j]/totales[2]
        elif i == 3:
            trans[i][j]=arr[i][j]/totales[3]
            
np.around(trans, decimals=2)
```
